```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

1. Implement smoothing filters

```python
from common.functions import convolution

def average_smooth(img, size):
    avg = 1 / size
    kernel = np.full((size, size), avg)
    return convolution(img, kernel)
    
def gaussian_smooth(img, size):
    avg = 1 / size
    kernel = np.full((size, size), avg)
    return convolution(img, kernel)
    
def median_smooth(img, size):
    return img
```

2. Eliminate some noises using either Gaussian filter and Median filter and compare their performance

```python
img = cv2.imread('images/01-lena-gray.png', cv2.IMREAD_GRAYSCALE)
sizes = (5, 11, 15)

img_gaussian = {}
img_med = {}

for size in sizes:
    img_gaussian[size] = gaussian_smooth(img, size)
    img_med[size] = median_smooth(img, size)

# Plot
plt.figure(figsize=(12, 6))

plt.subplot(1, 4, 1)
plt.title('original image')
plt.imshow(img, cmap='gray')

for i, size in enumerate(sizes):
    plt.subplot(2, 4, i + 2)
    plt.title(f'gaussian size {size}')
    plt.imshow(img_gaussian[size], cmap='gray')

    plt.subplot(2, 4, i + 6)
    plt.title(f'median size {size}')
    plt.imshow(img_med[size], cmap='gray')

plt.show()
```
