## 0. Image loading

1. Import library

```python
import cv2
from matplotlib import pyplot as plt
```

2. raw data

```python
raw = cv2.imread('images/pepper.raw', cv2.IMREAD_COLOR)

plt.imshow(cv2.cvtColor(raw, cv2.COLOR_BGR2RGB))
plt.show()
```

3. bmp data

```python
bmp = cv2.imread('images/baboon.bmp', cv2.IMREAD_COLOR)

plt.imshow(cv2.cvtColor(bmp, cv2.COLOR_BGR2RGB))
plt.show()
```

4. jpeg data

```python
jpeg = cv2.imread('images/goldhill.jpg', cv2.IMREAD_COLOR)

plt.imshow(cv2.cvtColor(jpeg, cv2.COLOR_BGR2RGB))
plt.show()
```
